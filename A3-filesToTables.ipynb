```python
#Positive-Books
# Parse the XML file -- 
tree = ET.parse(book_positive)
root = tree.getroot()

# Define a list to store extracted data
data = []

# Iterate through custom tags and extract data
for custom_tag in root.findall('review'):
    # Assuming your custom tag structure is like <custom_tag><field1>value1</field1>...</custom_tag>
    values = []
    for field in custom_tag:
        values.append(field.text)
    data.append(values)

# Define column names based on the fields you extracted
columns = ['unique_id', 'asin', 'product_name', 'product_type','helpful','rating', 'title', 'date', 'reviewer', 'reviewer_location', 'review_text']  # Adjust column names accordingly

# Create a DataFrame from the extracted data
df_books_positive = pd.DataFrame(data, columns=columns)

#formatted_output = df.head(10).to_string(index=False, justify='left')
#print(formatted_output)

#Print the first 10 rows as a formatted table
table = tabulate(df_books_positive.head(10), headers='keys', tablefmt='pretty')
print(table)
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

# Read the XML content as a string
with open(dvd_negative, 'r', encoding='utf-8') as xml_file:
    xml_content = xml_file.read()

# Remove the invalid character (ASCII value 26)
cleaned_xml_content = xml_content.replace('\x1A', '')

# Parse the cleaned XML content using ElementTree
root = ET.fromstring(cleaned_xml_content)

# Clean the XML data by removing special characters and non-ASCII characters
cleaned_data = []

for custom_tag in root.findall('review'):
    values = []
    for field in custom_tag:
        cleaned_text = field.text.strip()
        values.append(cleaned_text)
    cleaned_data.append(values)

# Define column names based on the fields you extracted
columns = ['unique_id', 'asin', 'product_name', 'product_type','helpful','rating', 'title', 'date', 'reviewer', 'reviewer_location', 'review_text']  # Adjust column names accordingly

# Create a DataFrame from the cleaned data
df_dvd_negative = pd.DataFrame(cleaned_data, columns=columns)

table = tabulate(df_dvd_negative.head(10), headers='keys', tablefmt='pretty')
print(table)

# Now you have your cleaned data in a DataFrame
print(df_dvd_negative)
```
